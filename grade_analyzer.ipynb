# Student Grade Analyzer
**Topic:** Python Function Definitions  
**Skills practiced:** `def`, parameters, return values, calling functions inside functions

---

## Step 1 — Create fake data with a function

**Concepts:** `def` keyword, `return` statement, dictionary  
This function takes NO parameters. It just creates and returns our dataset.

```python
def create_data():
    """
    Returns a dictionary of student names and their grades (0-100).
    """
    data = {
        "Alice":   [85, 92, 78, 95, 88],
        "Bob":     [72, 65, 70, 68, 74],
        "Carlos":  [90, 95, 92, 98, 94],
        "Diana":   [55, 60, 58, 62, 57],
        "Emma":    [80, 75, 83, 79, 81],
        "Frank":   [45, 50, 48, 52, 47],
        "Grace":   [99, 97, 95, 100, 98],
        "Henry":   [63, 68, 65, 70, 66],
    }
    return data


# Call the function and store the result
students = create_data()

print("Students loaded:", list(students.keys()))
print("Example - Alice's grades:", students["Alice"])
```

---
## Step 2 — Calculate average grade

**Concepts:** parameters, arithmetic operations, returning a computed value

```python
def calculate_average(grades):
    """
    Takes a list of grades and returns the average (float).
    
    Parameter:
        grades (list): a list of numeric scores
    
    Returns:
        float: the average score rounded to 2 decimals
    """
    total = sum(grades)
    count = len(grades)
    average = total / count
    return round(average, 2)


# Test it on one student
alice_avg = calculate_average(students["Alice"])
print(f"Alice's average: {alice_avg}")

# Now compute averages for ALL students using a loop
averages = {}
for name, grades in students.items():
    averages[name] = calculate_average(grades)

print("\nAll averages:")
for name, avg in averages.items():
    print(f"  {name}: {avg}")
```

---
## Step 3 — Assign letter grades

**Concepts:** if/elif/else, functions calling other functions, list comprehension

```python
def get_letter(score):
    """
    Converts a numeric score to a letter grade.
    
    Parameter:
        score (float): a numeric score from 0 to 100
    
    Returns:
        str: letter grade ('A', 'B', 'C', 'D', or 'F')
    """
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"


def assign_letters(averages_dict):
    """
    Takes a dict of {name: average_score} and returns a dict of {name: letter_grade}.
    Calls get_letter() internally — functions calling functions!
    """
    letters = {}
    for name, avg in averages_dict.items():
        letters[name] = get_letter(avg)   # <-- calling get_letter() here
    return letters


# Test get_letter alone first
print("Score 95 →", get_letter(95))
print("Score 72 →", get_letter(72))
print("Score 48 →", get_letter(48))

# Now assign for all students
letter_grades = assign_letters(averages)
print("\nLetter grades:")
for name, letter in letter_grades.items():
    print(f"  {name}: {letter}")
```

---
## Step 4 — Find top and bottom students

**Concepts:** `max()` / `min()` with `key=`, `dict.items()`, returning a tuple

```python
def get_top_student(averages_dict):
    """
    Returns the name and score of the highest-performing student.
    
    Returns:
        tuple: (name, score)
    """
    top_name = max(averages_dict, key=averages_dict.get)
    top_score = averages_dict[top_name]
    return (top_name, top_score)


def get_bottom_student(averages_dict):
    """
    Returns the name and score of the lowest-performing student.
    
    Returns:
        tuple: (name, score)
    """
    bottom_name = min(averages_dict, key=averages_dict.get)
    bottom_score = averages_dict[bottom_name]
    return (bottom_name, bottom_score)


top = get_top_student(averages)
bottom = get_bottom_student(averages)

print(f"Top student:    {top[0]} with average {top[1]}")
print(f"Bottom student: {bottom[0]} with average {bottom[1]}")
```

---
## Step 5 — Print a full report

**Concepts:** orchestration, f-strings, calling all previous functions together

```python
def print_report(data):
    """
    Master function: calls all previous functions and prints a complete report.
    
    Parameter:
        data (dict): raw student data from create_data()
    """
    # --- Compute everything using our functions ---
    avgs    = {name: calculate_average(grades) for name, grades in data.items()}
    letters = assign_letters(avgs)
    top     = get_top_student(avgs)
    bottom  = get_bottom_student(avgs)
    overall = calculate_average(list(avgs.values()))

    # --- Print the report ---
    print("=" * 45)
    print("        STUDENT GRADE REPORT")
    print("=" * 45)
    print(f"{'Name':<12} {'Average':>8} {'Grade':>6}")
    print("-" * 30)

    for name in data:
        print(f"{name:<12} {avgs[name]:>8.2f} {letters[name]:>6}")

    print("-" * 30)
    print(f"Class average:  {overall:.2f}")
    print(f"Top student:    {top[0]} ({top[1]})")
    print(f"Bottom student: {bottom[0]} ({bottom[1]})")
    print("=" * 45)


# --- RUN EVERYTHING ---
student_data = create_data()
print_report(student_data)
```
